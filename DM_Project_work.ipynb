```python
# zomato_scaled_new = StandardScaler().fit_transform(zomato_new.drop('rate',axis=1))
# zomato_scaled_new = pd.DataFrame(zomato_scaled_new, dtype=None, copy=False, index = zomato_new.drop('rate',axis=1).index, columns=zomato_new.drop('rate',axis=1).columns)
# zomato_scaled_new
```

```python
from sklearn.decomposition import PCA

pca = PCA()
principal = pca.fit_transform(zomato_scaled)

principal_component = pd.DataFrame(pca.components_.T , index = zomato_scaled.columns
             , columns = ['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
principalDf = pd.DataFrame(principal, index = zomato_scaled.index, columns = principal_component.columns)
principalDf
```

```python
principal_component
```

```python
principal_variance = pd.DataFrame(pca.explained_variance_ratio_*100 , columns = ['% Variance Preserved']
             , index = ['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13'])
principal_variance['Cumulative % Variance'] = principal_variance['% Variance Preserved'].cumsum()
principal_variance
```

```python
plt.figure(figsize=(13,6))
plt.plot(principal_variance.index, principal_variance['% Variance Preserved'], '-o')
plt.plot(principal_variance.index, principal_variance['Cumulative % Variance'], '-s')
plt.xticks(principal_variance.index)
plt.ylabel('Percentage of Variance Explained')
plt.xlabel('Principal Component')
plt.legend(['% Variance Preserved','Cumulative % Variance Preserved'])
plt.title('Variance Graph for PCs')

plt.show()
```

```python
import time
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


timet = np.zeros(3)
r2score = np.zeros(3)
rmse = np.zeros(3)

Xpca_Train, Xpca_Test, Ypca_Train, Ypca_Test = train_test_split(principalDf, y, test_size = 0.10, random_state = 0)

RForest1=RandomForestRegressor(n_estimators=50,min_samples_leaf=.0001)

start_time1 = time.time()
RForest1.fit(Xpca_Train,Ypca_Train)
end_time1 = time.time()

ypca_predict=RForest1.predict(Xpca_Test)

r2score[0] = r2_score(Ypca_Test,ypca_predict)
timet[0]=end_time1-start_time1
rmse[0]= mean_squared_error(Ypca_Test,ypca_predict,squared=False)
r2score[0]
```

```python
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(zomato_scaled, y, test_size = 0.10, random_state = 0)

start_time3 = time.time()
model = LinearRegression()

model.fit(x_train,y_train)
end_time3 = time.time()
y_l_pred = model.predict(x_test)
r2score[1] = r2_score(y_test,y_l_pred)
timet[1]=end_time3-start_time3
rmse[1]= mean_squared_error(y_test,y_l_pred,squared=False)
```
